```python
import pandas as pd
import numpy as np
```

```python
irsincomefilepath='resources/17incyallagi.csv'
```

```python
irsincome_df=pd.read_csv(irsincomefilepath)
```

```python
irsincome_df
```

```python
irsincome_df = irsincome_df[['STATE', 
                            'COUNTYNAME', 
                            'N1', 
                            'mars1', 
                            'MARS2', 
                            'ELDERLY', 
                            'A02650', 
                            'N02300', 
                            'A02300']]
irsincome_df
```

```python
irsincome_df = irsincome_df.rename(columns={'STATE': 'State', 
                                  'COUNTYNAME': 'County', 
                                  'N1': 'Number of Returns', 
                                  'mars1': 'Single Returns', 
                                  'MARS2': 'Joint Returns', 
                                  'ELDERLY': 'Elderly Returns', 
                                  'A02650': 'Total Income Amount', 
                                  'N02300': 'Returns w/ Uunemployment Compensation', 
                                  'A02300': 'Unemployment Compensation Amount'})
irsincome_df
```

```python
irsincome_df = irsincome_df.groupby('County').sum()
irsincome_df
```
