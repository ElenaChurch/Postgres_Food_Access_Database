```python
#importing stuff
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
```

```python
heartdiseasefilepath='resources/Heart Disease Mortality.csv'
foodaccessfilepath='resources/food_access.csv'
irsincomefilepath='resources/17incyallagi.csv'
```

```python
heartdisease_df=pd.read_csv(heartdiseasefilepath)
foodaccess_df=pd.read_csv(foodaccessfilepath)
irsincome_df=pd.read_csv(irsincomefilepath)
```

## cleaning of the income dataset 

```python
irsincome_df
```

```python
irsincome_df = irsincome_df[['STATE', 
                            'COUNTYNAME', 
                            'N1', 
                            'mars1', 
                            'MARS2', 
                            'ELDERLY', 
                            'A02650', 
                            'N02300', 
                            'A02300']]
irsincome_df
```

```python
irsincome_df = irsincome_df.rename(columns={'STATE': 'State', 
                                  'COUNTYNAME': 'County', 
                                  'N1': 'Number of Returns', 
                                  'mars1': 'Single Returns', 
                                  'MARS2': 'Joint Returns', 
                                  'ELDERLY': 'Elderly Returns', 
                                  'A02650': 'Total Income Amount', 
                                  'N02300': 'Returns w/ Uunemployment Compensation', 
                                  'A02300': 'Unemployment Compensation Amount'})
irsincome_df
```

```python
irsincome_df = irsincome_df.groupby('County').sum()
irsincome_df
```

## Cleaning of the heart dataset 

```python
heartdisease_df
```

```python
heartdisease_df=heartdisease_df[['Year','LocationAbbr','LocationDesc','GeographicLevel', 'Data_Value',
'Data_Value_Unit',
'Stratification1',
'Stratification2',
'LocationID'
]]
heartdisease_df
```

```python
overall_1 = heartdisease_df['Stratification1']=="Overall"
overall_2 = heartdisease_df['Stratification2']=="Overall"
county = heartdisease_df['GeographicLevel']=="County"

strat1 = heartdisease_df[overall_1]
strat2 = strat1[overall_2]
hd_df = strat2[county]
hd_df
```

```python
#finding count of NAN
hd_df.isna().sum()
#droping NAN
hd_df=hd_df.dropna()
#checking that it worked
hd_df.isna().sum()
```

```python
hd_df=hd_df.rename(columns={'Year': 'Year',
'LocationAbbr':'State',
'LocationDesc':'County',
'GeographicLevel':'Location Type',
'Data_Value':'Heart Disease Mortality Rate',
'Data_Value_Unit':'Value Unit',
'Stratification1':'Stratification 1',
'Stratification2':'Stratification 2',
'LocationID':'Location ID'})
hd_df
```

```python
clean_hd = hd_df[['State','County', 'Heart Disease Mortality Rate',
'Value Unit',
'Location ID'
]]
clean_hd
```

## Cleaning of the food access dataset 

```python
foodaccess_df=foodaccess_df[['County','State', 'LACCESS_POP15',
'PCT_LACCESS_POP15',
'LACCESS_LOWI15',
'PCT_LACCESS_LOWI15',
'LACCESS_HHNV15',
'PCT_LACCESS_HHNV15',
'LACCESS_SNAP15',
'PCT_LACCESS_SNAP15',
'LACCESS_CHILD15',
'PCT_LACCESS_CHILD15',
'LACCESS_SENIORS15',
'PCT_LACCESS_SENIORS15',
'LACCESS_WHITE15',
'PCT_LACCESS_WHITE15',
'LACCESS_BLACK15',
'PCT_LACCESS_BLACK15',
'LACCESS_HISP15',
'PCT_LACCESS_HISP15',
'LACCESS_NHASIAN15',
'PCT_LACCESS_NHASIAN15',
'LACCESS_NHPI15',
'PCT_LACCESS_NHPI15',
'LACCESS_MULTIR15',
'PCT_LACCESS_MULTIR15',
'LACCESS_NHNA15',
'PCT_LACCESS_NHNA15'
]]
foodaccess_df
```

```python
# rename columns for foodaccess_df
lowaccess_df=foodaccess_df.rename(columns={'County':'county','State':'state', 'LACCESS_POP15': 'population count',
'PCT_LACCESS_POP15':'population percent',
'LACCESS_LOWI15':'low income count',
'PCT_LACCESS_LOWI15':'low income percent',
'LACCESS_HHNV15':'households, no car count',
'PCT_LACCESS_HHNV15':'households, no car percent',
'LACCESS_SNAP15':'snap households count',
'PCT_LACCESS_SNAP15':'snap households percent',
'LACCESS_CHILD15':'children count',
'PCT_LACCESS_CHILD15':'children percent',
'LACCESS_SENIORS15':'seniors count',
'PCT_LACCESS_SENIORS15':'seniors percent',
'LACCESS_WHITE15':'white count',
'PCT_LACCESS_WHITE15':'white percent',
'LACCESS_BLACK15':'black count',
'PCT_LACCESS_BLACK15':'black percent',
'LACCESS_HISP15':'hispanic count',
'PCT_LACCESS_HISP15':'hispanic pecent',
'LACCESS_NHASIAN15':'asian count',
'PCT_LACCESS_NHASIAN15':'asian percent',
'LACCESS_NHPI15':'hawaiian or pacific islander count',
'PCT_LACCESS_NHPI15':'hawaiian or pacific islander percent',
'LACCESS_MULTIR15':'multiracial count',
'PCT_LACCESS_MULTIR15':'multiracial percent',
'LACCESS_NHNA15':'american indian or alaska native count',
'PCT_LACCESS_NHNA15':'american indian or alaska native percent'})
```

```python
lowaccess_df
```

```python
#cleaning lowaccess DF

# Change All the counts to whole #s
lowaccess_df['population count']=lowaccess_df['population count'].apply(np.floor)
lowaccess_df['low income count']=lowaccess_df['low income count'].apply(np.floor)
lowaccess_df['households, no car count']=lowaccess_df['households, no car count'].apply(np.floor)
lowaccess_df['households, no car count']=lowaccess_df['households, no car count'].apply(np.floor)
lowaccess_df['snap households count']=lowaccess_df['snap households count'].apply(np.floor)
lowaccess_df['hispanic count']=lowaccess_df['hispanic count'].apply(np.floor)
lowaccess_df['asian count']=lowaccess_df['asian count'].apply(np.floor)
lowaccess_df['asian count']=lowaccess_df['asian count'].apply(np.floor)
lowaccess_df['hawaiian or pacific islander count']=lowaccess_df['hawaiian or pacific islander count'].apply(np.floor)
lowaccess_df['multiracial count']=lowaccess_df['multiracial count'].apply(np.floor)
lowaccess_df['black count']=lowaccess_df['black count'].apply(np.floor)
lowaccess_df['children count']=lowaccess_df['children count'].apply(np.floor)
lowaccess_df['seniors count']=lowaccess_df['seniors count'].apply(np.floor)
lowaccess_df['hispanic count']=lowaccess_df['hispanic count'].apply(np.floor)
lowaccess_df['black count']=lowaccess_df['black count'].apply(np.floor)
lowaccess_df['american indian or alaska native count']=lowaccess_df['american indian or alaska native count'].apply(np.floor)

lowaccess_df
```
